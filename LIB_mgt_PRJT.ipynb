```python
class Library:
    def __init__(self):
        self.books = {
            "The Great Gatsby": {"author": "F. Scott Fitzgerald", "available": True},
            "1984": {"author": "George Orwell", "available": True},
            "Moby Dick": {"author": "Herman Melville", "available": True},
            "Hamlet": {"author": "William Shakespeare", "available": True},
            "Pride and Prejudice": {"author": "Jane Austen", "available": True},
            "To Kill a Mockingbird": {"author": "Harper Lee", "available": True},
            "The Catcher in the Rye": {"author": "J.D. Salinger", "available": True},
            "The Hobbit": {"author": "J.R.R. Tolkien", "available": True},
            "War and Peace": {"author": "Leo Tolstoy", "available": True},
            "Crime and Punishment": {"author": "Fyodor Dostoevsky", "available": True}
        }

    def display_books(self):
        print("Available Books:")
        for book, details in self.books.items():
            availability = "Available" if details["available"] else "Not Available"
            print(f" - {book} by {details['author']} [{availability}]")

    def borrow_book(self, book_name):
        if book_name in self.books:
            if self.books[book_name]["available"]:
                print(f"You have borrowed '{book_name}'")
                self.books[book_name]["available"] = False
            else:
                print(f"Sorry, '{book_name}' is already borrowed.")
        else:
            print(f"Sorry, '{book_name}' is not available in our library.")

    def return_book(self, book_name):
        if book_name in self.books:
            if not self.books[book_name]["available"]:
                print(f"Thank you for returning '{book_name}'")
                self.books[book_name]["available"] = True
            else:
                print(f"'{book_name}' was not borrowed.")
        else:
            print(f"'{book_name}' is not a valid book in our library.")

class Member:
    def choose_action(self):
        action = input("Would you like to borrow or return a book? (borrow/return): ").lower()
        return action

    def request_book(self):
        self.book = input("Enter the name of the book: ")
        return self.book

def main():
    library = Library()
    member = Member()

    while True:
        library.display_books()
        action = member.choose_action()

        if action == "borrow":
            requested_book = member.request_book()
            library.borrow_book(requested_book)
        elif action == "return":
            returned_book = member.request_book()
            library.return_book(returned_book)
        else:
            print("Invalid action! Please choose either 'borrow' or 'return'.")

        continue_choice = input("Do you want to continue? (yes/no): ").lower()
        if continue_choice == "no":
            print("Thank you for using the library system. Goodbye!")
            break

if __name__ == "__main__":
    main()
```
